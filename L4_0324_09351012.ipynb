```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df["age"].plot.hist(bins=30)
```

```python
df.groupby(["survived","sex"]).size().unstack(1)
```

```python
df.groupby(["survived","sex"]).size().unstack(1).plot(kind="bar")
```

```python
df.groupby(["survived","age"]).size().unstack(0).plot(kind="hist")
```

```python
df.boxplot(column="age",by="survived")
```

```python
sns.jointplot(x="age",y="survived", data=df)
```

```python
sns.kdeplot(x= "age", y = "survived", data = df)
```

```python
sns.pairplot(df)
```
